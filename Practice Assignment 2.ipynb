```python
import logging
logging.basicConfig(filename = "q1.log", level = logging.DEBUG , filemode = 'w', format = "%(asctime)s:%(levelname)s:%(message)s")


class listSearch:
    
    def __init__(self,data):
        logging.info("List object is initialised")
        self.data = data
        
    def search(self, value):
        logging.info("Searching...")
        try:
            for i in range(len(self.data)):
                if self.data[i] == value:
                    logging.info("Successful search")
                    return i, True
            else:
                logging.info("Unsuccessful search")
                return False
                
        except Exception as e:
            logging.error("Some error occurred")
            logging.error(e)
        
data = [1,2,3,4,56,4,2,5]
lobj = listSearch(data)
lobj.search(6)
```

```python
import logging
import pymongo
logging.basicConfig(filename = "mongo.log", level = logging.DEBUG , filemode = 'w', format = "%(asctime)s:%(levelname)s:%(message)s")

class CRUDmongo:
    def __init__(self,name,password):
        self.name = name
        self.password = password
    
    def connect(self):
        try:
            self.client = pymongo.MongoClient(f"mongodb+srv://{self.name}:{self.password}@cluster1.vrhbe8i.mongodb.net/?retryWrites=true&w=majority")
            logging.info("mongodb connected successfully")
        except Exception as e:
            logging.error("some error occured while connection")
            logging.error(e)
    
    def createdb(self,dbname):
        try:
            self.dbname = dbname
            self.db = self.client[dbname]
            logging.info("database created successfully")
        except Exception as e:
            logging.error("some error occured while creating database")
            logging.error(e)
    def create_collection(self,cname):
        try:
            self.cname = cname
            self.coll = self.db[f"{self.cname}"]
            logging.info("collection created successfully")
        except Exception as e:
            logging.error("some error occured while creating collection")
            logging.error(e)
    def insert_data(self):
        try:
            print("give the inputs")
            idi = int(input("Enter ID: "))
            fname = input("Enter the firstname: ")
            lname = input("Enter the last name: ")

            data = {
                "idi" : idi,
                "fname": fname,
                "lname": lname

            }
            logging.info(f"data {data} captured successfully")
            self.coll.insert_one(data)
            logging.info("Data inserted successfully")

        except Exception as e:
            logging.error("Some error occured during insertion")
            logging.error(e)

    def read_data(self):
        try:
            logging.info("Reading data......")
            print("####### THE DATA #########")
            records = self.coll.find()
            for i in records:
                print(i, sep = ",")
            logging.info("fetched data successfully")

        except Exception as e:
            logging.error("some error occurred during reading data")
            logging.error(e)

    def update_data(self):
        try:
            logging.info("updating started")
            idi = int(input("Enter ID To Update Data: "))
            fnm = input("Enter new first name: ")
            lnm = input("Enter the last name: ")

            rec = self.coll.find({"idi":idi})
            if rec:
                self.coll.update_one({"idi":idi},{"$set":{"fname":fnm, "lname":lnm }})
                logging.info("Data updated successfully")
            
        except Exception as e:
            logging.error("some error occured while updating")
            logging.error(e)
            
    def delete_data(self):
        try:
            logging.info("deletion started")
            idi = int(input("Enter the id to be deleted"))
            
            re = self.coll.find({"idi":idi})
            
            if re:
                print(re)
                
                self.coll.delete_one({"idi":idi})
                logging.info("deleted successfully")
                
        except Exception as e:
            logging.error(e)
        


cobj = CRUDmongo('karthik', 'karthik')

cobj.connect()
cobj.createdb('ineuron')
cobj.create_collection("coll")
cobj.insert_data()
cobj.update_data()
cobj.delete_data()
```

```python
logging.basicConfig(filename='record_from_mongo.log', level=logging.DEBUG , filemode='w', format="%(asctime)s %(levelname)s %(message)s")

class Import_from_Mongo():
    
    def __init__(self, clienturl, filename):
        self.clienturl = clienturl
        self.filename = filename
        try:
            self.client = pymongo.MongoClient(self.clienturl)
        except Exception as e:
            print("Connection not established", e)
            logging.error("Connection not established", e)
        else:
            print("Connection to MongoDB is successful")
            logging.info("Connection to MongoDB is successful")
            
    def connect_to_collection(self, db_name, collection_name):
        try:
            self.database = self.client[db_name]
            self.collection = self.database[collection_name]
        except Exception as e:
            print("Connection not established", e)
            logging.error("Connection not established", e)
        else:
            print("Connection to collection is successful")
            logging.info("Connection to collection is successful")
    
    def importing_record(self):
        try:
            record = []
            for i in self.collection.find():
                    record.append(i)
            with open(self.filename, 'a') as f:
                f.write(str(record))
        except Exception as e:
            print("Importing unsuccessful", e)
            logging.error("Importing unsuccessfull", e)
        else:
            print("Importing successful")
            logging.info("Importing successfull")
            
    def show_data(self):
        try:
            with open(self.filename, 'r') as f:
                record = f.read()
                print(record)
                logging.info("Reading operation is successful")
        except Exception as e:
            print("There is some error", e)
            logging.error("There is some error", e)
```
